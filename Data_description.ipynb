```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# ðŸ“Œ 1ï¸âƒ£ Daten laden
df = pd.read_csv("C:/Users/Nael Ackle/OneDrive - BBBaden/Dokumente/IAP/Modul 259/main.csv")

print("\nðŸ“Š Statistische Kennzahlen:")

# ðŸ“Œ 2ï¸âƒ£ Spaltennamen definieren
relevant_columns = ['Open', 'Close', 'High', 'Low', 'Volume']

for col in relevant_columns:
    print(f"\n--- {col.capitalize()} ---")  # Spaltennamen mit grossem Anfangsbuchstaben
    print(f"Mittelwert:            {df[col].mean():.6f}")
    print(f"Median:                {df[col].median():.6f}")
    print(f"Modus:                 {df[col].mode().iloc[0]:.6f}")
    print(f"Standardabweichung:    {df[col].std():.6f}")
    print(f"Varianz:               {df[col].var():.6f}")
    print(f"25%-Perzentil:         {df[col].quantile(0.25):.6f}")
    print(f"50%-Perzentil (Median):{df[col].quantile(0.50):.6f}")
    print(f"75%-Perzentil:         {df[col].quantile(0.75):.6f}")

# ðŸ“Œ 3ï¸âƒ£ Visualisierung der Daten

plt.figure(figsize=(15, 8))  # GrÃ¶ssere Darstellung fÃ¼r bessere Sichtbarkeit

for i, col in enumerate(relevant_columns, 1):
    plt.subplot(2, 3, i)  # Erzeugt eine Rasterdarstellung mit 2 Zeilen, 3 Spalten
    sns.histplot(df[col], bins=50, kde=True, color="blue")
    plt.xlabel(col)
    plt.ylabel("HÃ¤ufigkeit")
    plt.title(f"Histogramm von {col}")

plt.tight_layout()  # Vermeidet Ãœberlappungen
plt.show()

# ðŸ“Œ 4ï¸âƒ£ Boxplots fÃ¼r eine bessere Ãœbersicht
plt.figure(figsize=(12, 6))
sns.boxplot(data=df[relevant_columns], palette="coolwarm")
plt.title("Boxplot der ausgewÃ¤hlten Spalten")
plt.xlabel("Spalten")
plt.ylabel("Werte")
plt.show()
```

Der Schlusskurs ist das wichtigste Merkmal fÃ¼r Vorhersagen, da er die letzte Bewertung eines Zeitintervalls darstellt.
Trader nutzen ihn oft, um Trends zu erkennen und kÃ¼nftige Preise vorherzusagen.


```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler, StandardScaler

# ðŸ“Œ 1ï¸âƒ£ CSV-Datei laden
df = pd.read_csv("C:/Users/Nael Ackle/OneDrive - BBBaden/Dokumente/IAP/Modul 259/main.csv")

# ðŸ“Œ 2ï¸âƒ£ Relevante Spalten definieren
columns_to_scale = ["Open", "Close", "High", "Low", "Volume"]

# ðŸ“Œ 3ï¸âƒ£ MinMaxScaler (Skalierung zwischen 0 und 1)
minmax_scaler = MinMaxScaler()
df_minmax_scaled = df.copy()  # Originaldaten behalten
df_minmax_scaled[columns_to_scale] = minmax_scaler.fit_transform(df[columns_to_scale])

print("\nâœ… MinMax-Skalierte Daten (0 bis 1):")
print(df_minmax_scaled.head())

# ðŸ“Œ 4ï¸âƒ£ StandardScaler (Mittelwert = 0, Standardabweichung = 1)
standard_scaler = StandardScaler()
df_standard_scaled = df.copy()
df_standard_scaled[columns_to_scale] = standard_scaler.fit_transform(df[columns_to_scale])

print("\nâœ… Standard-Skalierte Daten (Mittelwert = 0, Std-Abweichung = 1):")
print(df_standard_scaled.head())
```
